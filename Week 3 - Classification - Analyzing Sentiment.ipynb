#### Problems with threshold classifier
* How to list positive / negative words ?
* Words have different degrees of sentiment: How do we weigh different words?
* Single words are not good enough: E.g : good -> positive ; not good -> negative;

#### The first two issues can be resolved by learning a classifier;

### Linear Classifier
* It takes all the words and adds weights to them.
* Example

                Words                Weights
                ----------------------------
                good                   1.0
                great                  1.5
                awesome                2.7
                bad                   -1.0  
                terrible              -2.1
                aweful                -3.3
                restaurant,we,the,etc  0.0

* Scoring a sentence.
* Example:
* Input x -> Sushi was "great" , the food was "awesome", but the service was "terrible".
* Score (x) = 1.5 + 2.7 - 2.1 = 2.1 
* Score(x) > 0 = positive review;
* Output is the weight of the input

                    |            WORDS       |          WEIGHTS         |
                     ----------------------------------------------------
                                             |
                                            \ /
                     ----------------------------------------------------                      
                     |               Simple LINEAR classifier           |
                     | * Score(x)= weighted count of words in sentence. |
                     | * If Score(x) > 0 :                              |
                     |      output (y hat) = positive;                  |
      Sentence       |   else                                           |
      from     --->  |       output (y hat) = negative;                 |
      review         |                                                  |
      (input x)      |                                                  |
                     |                                                  |
                     |                                                  |        
                     |--------------------------------------------------|
                     

### Decision Boundary:
* Classifier attempt to make decisions.
* Decision boundaries : Score(x) = weight of word1 * # occurence of the word1 + weight of word2 * # occurence of the word2 + ...
* A plot based of the various words and their ocuurence can be visualized.
* A prediction can be made on the plot an dline segregates the positive from the negative reviews.
* The classifier / decision boundary / line of segregation is 0 and therefore is termed the linear classifier, linear decision boundary.

#### For linear classifiers:
* When 2 weights are non-zero -> line
* When 3 weights are non-zero -> plane (positive prediction are above the plane ; negative predictions are below the plane);
* When many weights are non-zer0 -> hyperplanes

##### For more generale classifiers (not linear) - more complicated shapes - segregate the data.

### Evaluation - Classification error:
* Testexample -> Learned classifier -> output predicted ; test error - evalutes the predicted output w.r.t true label -> if its a match then its correct else it is false;

#### Examples
 * Review (Sushi was great, +) -> sentence : Sushi was great, true label : +;
 * the sentence "Sushi was great" is feed to the learned classifier; while he true label "+" is hidden;
 * The classifier predicts the output. In this case -> +
 * Since the predicted output = + , the true label = + ; its a match - therefore Correct;
 * -------------------
 * Review (Food was OK, -) -> Sentence : Food was OK, true label : -;
 * the sentence "Food was ok" is feed to the learned classifier; while the true label "-" is hidden;
 * the classifier predicts the output. In this case -> +
 * Since the predicted output = + , the true label = - ; its a mis-match - therefore Mistake;
 
##### The process is repeated for all examples.
 
### Classification Error and  Accuracy:
 * Error measures the fraction of mistake  = # mistakes / total # of sentences;
 * Best possible value of error = 0.0;
 * error = 1 - accuracy;
 
 * Accuracy  measure the fraction of correct predictions = # correct / total # of sentences;
 * Best possible value of accuracy = 1.0;
 * accuracy = 1 - error;
 
#### For Good Accuracy -> prevent random guessing
* for binary classification - on average - accuracy = 0.5 ,  rest random guessing;
* for k class, accuracy = 1/k ; 0.333 for 3 classes, 0.25  for 4 classes, etc;

#### Is a classifier with 90% accuracy good? 
* Depends
* In case of spam - not good - case of major class prediction - It has amazing performance when there is a "class imbalance" (One class is more common than another, beats random (if you know the majority class));

### Learning Curves

#### How much data does a model need to learn?
* The more the better - but data quality is an important factor.
* Theoretical techniques can bound the amount of data needed - typically loose for practical application, but provides guidance.
* In practice - more complex models require more data; empirical analysis can provide guidance.

#### Learning curves - relates the amount of training data and the test error.
* If the amount of training data is high, then the test error will be low ;
* If the amount of training data is low , then the test error will be high;
* Each model has a gap - bias, beyond which the although the number of training data increases the test error will not decrease.

#### More complex models tend to have less bias.
* Sentiment classifier using single words can do Ok;
* Less bias -> potentially more accurate, needs more data to learn.

##### Models with less bias tends to need more data to learn well, but to do better with sufficient data.

### Create a word count vector for each review

### Build a sentiment classifier

```python
products['rating'].show(view='Categorical')
```

### Defining positive (4,5) and negative (1,2) reviews; 3 is removed -> data engineering

```python
products = products[products['rating'] != 3]
```

```python
products['sentiment'] = products['rating'] >=4
```

```python
products.head()
```

### Training the sentiment classifier

```python
train_data, test_data = products.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data, 
                                                      target='sentiment', 
                                                      features = ['word_count'], 
                                                      validation_set=test_data)
```

### Evaluate the sentiment model

```python
sentiment_model.evaluate(test_data, metric = 'roc_curve') # roc_curve - explore confusion matric, false positive, false negatives;
```

```python
sentiment_model.show(view='Evaluation')
```

### Apply the learned model to understand the product giraffe

```python
giraffe_reviews['predicted_sentiment'] = sentiment_model.predict(giraffe_reviews, output_type='probability')
```

```python
giraffe_reviews.head()
```

### 1. Use .apply() to build a new feature with the counts for each of the selected_words

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def word_count(word, word_dict):
    if word in word_dict:
        return word_dict[word];
    else:
        return 0;
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda word_dict : word_count(word, word_dict))
```

```python
products.head()
```

```python
sumlist = []
for word in selected_words:
    temp = products[word].sum()
    print "%s : %i" %(word, products[word].sum())
    sumlist.append(temp)
```
