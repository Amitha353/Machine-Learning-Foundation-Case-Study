### Applications of Deep Learning to Computer Vision
* Image features = local detectors, combined to make predictions;
* Local detectors look for "interesting points" in the image;
* Image features - collections of locally interesting points combined to build classifiers;
* Many hand created features exist for finding interest points...
    1. Spin Images
    2. Textons
    3. RIFT
    4. GLOH
    5. HoG
    6. SIFT

#### Standard image classification approach
* Using the hand created image feature - SIFT  feature;
* Input   -> Extract features (create a vector - based on the locations where the firing occured the value exists in the vector)  -> feed it to a simple classifier (e.g., logistioc regression, SVMs) [detect]; 

#### Challenge - Many hand created features exist for finding interest points - but it is a painful process to design.

### Challenges of Deep Learning
#### Pros
* Enables learning of features rather than hand tuning;
* Impressive performance gains for - Computer Vision, Speech recognition, some text analysis;

#### Challenges
* Requires a lot of data, and labelled data;
* Split into training / validation set -> train on deep neural network -> takes time;
* Computationally expensive, extremely hard to tune;
* Extremely hard to tune
    * Choice of architecture
    * Parameter types
    * Hyperparameters
    * Learning algorithms,
       etc...

### Image Classification with Deep Features

```python
import graphlab
```

#### Load common image analysis dataset

```python
# Since the dataset is already segregated into train and test set; 
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')
```

#### Exploring the image data

```python
graphlab.canvas.set_target('ipynb')
```

```python
image_train.head()
```

```python
image_train['image'].show()
```

#### Train classifier on the raw image pixels (Features -> raw pixels and not deep features) 

```python
raw_pixel_model = graphlab.logistic_classifier.create(image_train, 
                                               target='label', 
                                               features=['image_array'])
```

#### Make prediction with the simple model based on raw pixels

```python
image_test[0:3]['image'].show()
```

```python
image_test[0:3]['label']
```

```python
raw_pixel_model.predict(image_test[0:3])
```

#### Evaluating raw pixel model on test-data

```python
### Evaluate - computes error metrics.
raw_pixel_model.evaluate(image_test)
```

#### The accuracy of the above model is veru low - 48%

#### Improving the model using the Deep Features (Using Transfer Learning)

```python
len(image_train)
```

#### Given the deep features, lets train a classifier

```python
deep_feaures_model = graphlab.logistic_classifier.create(image_train,
                                                        features=['deep_features'],
                                                        target='label')
```

#### Apply the deep features model to the first few images in the test set

```python
image_test[0:3]['image'].show()
```

```python
deep_feaures_model.predict(image_test[0:3])
```

#### Compute the test data accuracy of the deep_feature_model

```python
deep_feaures_model.evaluate(image_test)
```

#### Accuracy of the Deep features model is better - Accuracy = 79%
