### Q7 : In what range is the accuracy of the 1-nearest neighbor classifier at classifying ‘dog’ images from the test set? 

```python
image_test_automobile = image_test[image_test['label'] == 'auomobile']
image_test_dog = image_test[image_test['label'] == 'dog']
image_test_cat = image_test[image_test['label'] == 'cat']
image_test_bird = image_test[image_test['label'] == 'bird']
```

```python
dog_automobile_neighbors = knn_model_automobile.query(image_test_dog, k =1)
dog_dog_neighbors = knn_model_dog.query(image_test_dog, k =1)
dog_cat_neighbors = knn_model_cat.query(image_test_dog, k =1)
dog_bird_neighbors = knn_model_bird.query(image_test_dog, k =1)
```

```python
dog_distance = graphlab.SFrame({'dog-automobile-distance': dog_automobile_neighbors['distance'],
                               'dog-dog-distance': dog_dog_neighbors['distance'],
                               'dog-cat-distance':dog_cat_neighbors['distance'],
                               'dog-bird-distance':dog_bird_neighbors['distance']})
```

```python
dog_distance.head()
```

```python
def is_dog_correct(row):
    if (row['dog-dog-distance'] < row['dog-automobile-distance'] and 
        row['dog-dog-distance'] < row['dog-cat-distance'] and row['dog-dog-distance'] < row['dog-bird-distance']):
        return 1
    else:
        return 0
```

```python
is_dog_correct(dog_distance[0])
```

```python
dog_distance.apply(is_dog_correct).sum()
```

```python
dog_distance.num_rows()
```

```python
total = float(dog_distance.num_rows())
dog_correct = float(dog_distance.apply(is_dog_correct).sum())
Result = float(dog_correct / total)
print Result * 100,'%'
```

### Result : 67.8 % (60 - 70)
