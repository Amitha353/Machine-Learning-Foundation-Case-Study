### Building a recommender system 
#### 1. Solution : Popularity (of each product);
* New articles -> most read , most emailed, etc;
##### Limitation -> No personalization, the result is based on the entire set of readers and not an individual;

#### 2. Solution : Classification Model
* The model will be use to evaluate whether a user likes or dislikes a product.
      
      User Info                              |-----> Yes!
      Purchase History  -----> Classifier  --
      Product info                           |-----> No
      Other info
      
##### Pros
* Personalized: Consider user info and purchase history, ..
* Features can capture context : Time of the day, what I just saw, ..
* Even handles limited user history : Age of user, etc, ..

##### Limitation
* Features may not be available; (user info, age, gender, prefernce, etc, product info); Since the model is dependent on these features it could be an issue;
* Often doesn't perform as well as collaborative filtering methods.

#### 3. Solution : People who bought this also bought ... (Collaborative Filtering)
#### Co-occurence matrix
* People who bought a phone also bought phone cover;
* Matrix C:It stores # users who bought both items i and j;
 - (# items x # items) matrix
                                ___                 ____
                      ___|_____|   |_______________|    |____________ 
                phone(i__|_____|pp |_______________|ppc |____________)
             C  =     t  |     |   |               |    |
                      e  |     |   |               |    |
                      m__|_____|   |_______________|    |_____________ 
         phone cover (s________|pcp|_______________|pcpc|_____________)  
                         |     |   |               |    |
                         |     |   |               |    | 
                         |_____|___|_______________|____|_____________
                               i t e m s
                              phone              phone cover
                               
                        pp -> # phone purchased
                        pcp = ppc # phone - phone cover purchases that occured;      
 - symmetric : # purchasing i and j same as # for j & i (Cij = Cji);
 ##### The diagonal is a reflection of values across the diagonal;
 * The co-occurence matrix will be built by iterating/searching through all the users history of purchases that have been made and count. Incremting the marix every time the required product is found;
 
### Making recommendations using co-occurences
#### User has purchased a phone;
#### Now to make recommendations to the user the following process is helpful.
1. Look at the "phone" row of the matrix (shown above) and extract it (as shown below - product and the count of the purchases of that product);
        [   | |  1  | |  |     10     |   |   2   |  | ...]
               phone       phone cover      charger

2. Recommendations can then be made, by sorting the above listed vector and recommend the items with the largest count.
 - phone cover, charger, etc;

#### Issue with Co-occurence matrix is that it must be normalized.
* What if there are very popular items?
- Popular electronic item : Phone;
- For any electronic item (eg :- Phone cover ) (The recommendation regardless of what has been purchased the recommendation will be a phone since it has a large count Cij for j=Phone, rather than recommending other Phone covers);
- Recommendations based on popularity drowns out other effects. (Needs to be avoided);

### Normalize co-occurences : Similarity Matrix
* Jaccard Similarity : normalizes by popularity; similar to TF-IDF (Term Frequency - Inverse Document Frequency);
- Who purchased i and j divided by i or j;
                # purchased by i and j
                ----------------------
                # purchased by i or j
* Many other similarity metrics possible, e.g., cosine similarity;

#### Limitation of Normalizing co-occurence
* Only the current page matters, no history: Recommend similar items to the one you bought and no purchase history considered;

#### (Weighted) Average of purchased items
* Consider a user has bought items {phone, phone cover}
- Compute user-specific score for each item j (charger, bluetooth, headset, earphone, etc)in the inventory by combining similarities:

* Score(User, charger) = 1/2 (S charger,phone + S charger,phonecover);
- A weighted average is computed based on having purchased a phone and a phone cover before (purchase history);

        Phone row vector         [   | |  1  | |  |     10     |   |   2   |  | ...]
                                        phone       phone cover      charger
                                   
        Phone cover row vector   [   | |  1  | |  |     10     |   |   2   |  | ...]                                   
                                        phone       phone cover      charger
                                                                                                                                                                                                                                 
                                        





* The row for phone is searched and the number of times a charger is purchased is noted;
* The row for phone cover is searched and yhe number of times a charger is purchased is noted;
* The results are averaged -> to figure out how likely it is for the user to purchase a charger given that they have already purchased a phone and phone cover;

* Could also weight recent purchased more heavily;
##### Sort Score(user, j) and find item j with the highest similarity.

#### Limitations:
* Does not use
 - context (e.g., time of day);
 - user features (e.g., age); - considers all users;
 - product features (e.g., clothes vs electronics) - considers the whole pool of users;
* Cold start problem:
 - In case of new users, new products, how can recommendations be made?
 - Since there is no purchase history for a new user;
 - Since there is no purchase history for a new product;

#### Solution 3: Discovering hidden structure by matrix factorization
* So far the features of the users nor the features of the product are not taken into considerstion, rather the recommendation is driven by counting the co-occurences of purchases and user histories;
* Instead the natural approach is to take into consideration, the person, their features and the products features and find a sync;
* It is also important to take into consideration the interaction between the users and the products;

### Movie Recommendation
* Table : Users watch movies and rate them.
<img src='User_movies_rating.png'>
- Each user only watches certain number of movies from the total set of movies;

#### Matrix completion problem
* The user movie rating table shown above will be transformed into the matrix shown below.
<img src='matrix_completion_problem.png'>
* There are lots of users, movies and the data is very sparse since only a subset of the movies are watched;
* White Squares -> Rating(u,v) - is unknown;
* Black cells -> Rating(u,v) - is known;
* Goal :- Fill the missing data (White cells),by using all the existing information (black cells);

### Recommendations from known user/item features
* Describe movie v with topic Rv (How much action, drama, romance, mystery, ...)
- Example : Kill Bill  [ 1.8  | 1.5 |  0.4  |  0.7  | ... ] <- Rv
                        action drama romance mystery  ... 
* Describe user u with topics Lu (How much the user likes action, drama, romance, mystery, ...)
- Example : User 1 - [ 1.2   | 1.2  | 0.2    | 0.5  | ... ]  <- Lu
                      action  drama  romance  mystery  ...

* Rating(u,v) (hat) - is the product of two vectors. -hat -> estimate, here R(u,v)hat - is the estimate of how much the user 'u' will like a movie 'v' they have never seen before.

       [ 1.8  | 1.5 |  0.4  |  0.7  | ... ]
       [ 1.2   | 1.2  | 0.2    | 0.5  | ... ]

       Result = 1.8 * 1.2 + 1.5 * 1.2 + 0.4 * 0.2 + 0.7 * 0.5 + ...;
        
* Recommendations - Sort movies users haven't watched by Rating(u,v) (hat) [highest to lowest];
* There is a high sync in the rating then the respective movies are recommended to the user;

### Matric factorization model : Discovering topics from data.
* Here intead an the estimation model is used, where the missing data values are filled by deriving the information from known cells;
<img src="Matrix_factorization_model.png">
* Only use observed values to estimate "topic" vectors Lu hat and Rv hat.
* Use the estimated Lu hat and Rv hat for recommendations.

* The fit on the model is the Residual Sum of Squares (RSS);
                RSS(L, R)= (Observerd (L, R) - Predicted(L, R))^2 + include all (u, v) pairs where Rating Observered(u,v) is available (black squares));
                
#### Limitation of Matrix factorization:
* Cold-start problem: This model still cannot handle new users or movies.
<img src='limitation_matrix_factorization.png'>

### Featurized Matrix factorization:
* Combining features and discovered topics.
* In case of a new user, using their features like age, gender, preferences, etc, predict items to the user;
* Moreover, as more information about the user is discovered matrix factorization topics become more relevant;

### Performance metric for recommender systems
* Precision = # liked and shown / # shown; (How much thing i have to look into compared to what i like);
* Recall = # likes and shown / # liked; (How much the recommended items cover the things that i'm interested in)

### Optimal recommenders
* Maximize Recall -> like all the recommended products, but the precision - small (liked items can be a small fraction of all the recommended products);
* Optimal Recommenders -> One that recommends all and only products like by the user; Here the precision and recall are both 1;

### Precision - recall curve
* Input - A specific recommender system
* Output - Algorithm-specific precision-recall curve;
<img src='precision-recall-curve.png'>
* the curve is drawn by varying the threshold on # items recommended.

### Which Algorithm is Best?
* For a given precision, want recall as large as possible (or vice-versa);
* One metric : largest area under the curve (AUC);
* Alternative : set desired recall and maximize precision;

#### Count number of users

```python
users = song_data['user_id'].unique() # get all the unique users
len(users)
```

#### Create a song recommender

```python
train_data, test_data = song_data.random_split(.8, seed=0)
```

### Simple popularity-based recommender

```python
popularity_model = graphlab.popularity_recommender.create(train_data, 
                                                          user_id='user_id',
                                                         item_id='song')
```

#### Use the popularity model to make predictions

```python
popularity_model.recommend(users=[users[0]])  # Recommending songs for user 0 based on popularity model.
```

```python
popularity_model.recommend(users=[users[1]])  ## Recommendtion for user 2 
```

#### Based on the popularity model, all users are recommended the same items.

### Build a song recommender with personalization

```python
personalized_model = graphlab.item_similarity_recommender.create(train_data,
                                                                user_id='user_id',
                                                                item_id='song')
```

#### Applying personalized model to make song recommendation

```python
personalized_model.recommend(users=[users[0]])
```

```python
personalized_model.recommend(users=[users[1]])
```

#### The result are personalized to the person.
#### Recommending song to the users

```python
personalized_model.get_similar_items(['With Or Without You - U2']) ## All the song similar to the one you choose.
```

```python
personalized_model.get_similar_items(['Chan Chan (Live) - Buena Vista Social Club'])
```

### Quantitative comparision between models - using precision and recall


```python
%matplotlib inline 
model_performance = graphlab.recommender.util.compare_models(test_data,
                                                            [popularity_model,personalized_model],
                                                            user_sample=0.05)
```

### Scenaro -3 :
* Split the data into 80% training, 20% testing, using seed=0, as was done in the iPython notebook above.
* Train an item_similarity_recommender, as done in the iPython notebook, using the training data.
* Next, we are going to make recommendations for the users in the test data, but there are over 200,000 users (58,628 unique users) in the test set. Computing recommendations for these many users can be slow in some computers. Thus, we will use only the first 10,000 users only in this question.
* Compute one recommended song for each of these test users.
* use .groupby() to find the most recommended song, use the COUNT aggregator instead of SUM, and store the results in a column we will call ‘count’.

```python
song_data.head()
```

```python
train_data_set, test_data_set = song_data.random_split(.8, seed=0)
```

```python
personalized_model_assignment = graphlab.item_similarity_recommender.create(train_data_set,
                                                                    user_id='user_id',
                                                                    item_id='song')
```

```python
subset_test_users = test_data_set['user_id'].unique()[0:10000]
```

```python
subset_test_model = personalized_model_assignment.recommend(subset_test_users, k = 1)
```

```python
song_count = subset_test_model.groupby(key_columns='song', operations={'count': graphlab.aggregate.COUNT()})
```

```python
song_count.sort('count', ascending=False)
```
